```python
#Bag of words 

# step 1 : lower all the sentences, removing special characters using re and  removing stop words 
# step 2 : Taking vocubulary, ie unique words and find its frequency. and order it in descending. more repeated word in first. By using "max_features" option in python to select required features.
# step 3 : Creating vector for each sentence but in ohe for each word we create a vector, that is the difference.  
```

```python
# Eg: 
# s1 : The food is good.
# s2 : The food is not good.
# s3 : The food is good and bad.

# stop words are : the, is , and 
# 1 - food good
# 2 - food not good 
# 3 - food good bad

# unique words: 
# food  good  not  bad 

# food 3 
# good 3
# not 1
# bad 1 

# vector is : 
#             [food,good,not,bad]
# food good  - [1,1,0,0]
# food not  good - [1,1,1,0]
# food bad good - [1,1,0,1]

# 2 types of BOW : 
#   - Binary BOW - vectors only contains 0,1 
#   - Normal BOW  - vectors may contains 0,1,2,3... 

#    Normal BOW Eg : 
#     food good food - [2,1,0,0]


# Advantages of BOW:
# - Easy to implement and intuitive. 
# - fixed size of input but not in ohe. 

# Disadvantages :
# - sparse matrix leads to overfitting if features increases 
# - Ordering of the words is changed. Bcz we are indexing the features based on frequency.
#     [food good not bad ] --> order 
#     S - food bad good -> [1,1,0,1] --> same vector but order of the word is changed. meaning is completely changed.
# - out of vocubulary
# - semantic meaning is not captured.
#    s1 - The food is good 
#    s2 - The food is not good. 
#    unique - The food is good not 
#    s1 - [1,1,1,1,0]
#    s2 - [1,1,1,1,1]
   
#    if we plot the vectors in 3d plane , the distance or angle will be very less by using cosine similarity due to only one point is varied. but the meaning is completely different. 


```

```python
ds = pd.read_csv('spam.csv',encoding='latin1')
print(ds.iloc[5549])
ds
```

```python
ds.drop(columns=["Unnamed: 2","Unnamed: 3","Unnamed: 4"],inplace=True) # inplace ==> True - no need to reassign the ds , after delete it will affect in ds
ds
```

```python
import re
import nltk

from nltk.corpus import stopwords

from nltk.stem import WordNetLemmatizer

lemmatiser = WordNetLemmatizer()

nltk.download('all')

```

```python
stop_words = stopwords.words('english')
stopwords.words()
```

```python
ds_message = ds["v2"]
ds_label = ds["v1"]
print (ds_message)
print (ds_label)
```

```python


for i in range(len(ds_message)):
    
    sentence = ds_message[i]
    s = re.sub("[^A-Za-z0-9]"," ",sentence)
    ss = s.lower().split()
    word = [lemmatiser.lemmatize(word,pos='v') for word in ss if word not in stopwords.words() ]
    w1 = " ".join(word)
    ds_message[i] = w1


ds_message
   
```

```python
New_dataframe = pd.concat([ds_message,ds_label],axis=True)
New_dataframe.rename(columns={'v1': 'labels', 'v2': 'messages'},inplace=True)
New_dataframe
```

```python
### Create Bag of words
# sklearn for Bag of words 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,binary=True,ngram_range=(2,2)) #binary ==> if true frequency will be a one or zero if false its frequency
a = cv.fit_transform(New_dataframe['messages']).toarray()
a
```

```python
# important parameters used in CountVectoriser 

# max features = Most repeated vocubulary as features
# binary =if true frequency will be a one or zero if false its frequency
# ngrams = bigrams, trigrams (combining 2 or 3 consecutive words )
# stopwords 
```
